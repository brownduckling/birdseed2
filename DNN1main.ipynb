```python
from keras.layers import Dense, Input
from keras.models import Sequential,Model
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1=np.random.rand(10)*.2+.4
x2=np.random.rand(10)*.2+.7
print('x1=',x1)
print('x2=',x2)
```

```python
x3=np.array([np.r_[x1,x2]])
print(x3)
```

```python
x5=x3.T
```

```python
model1=Sequential()
model1.add(Dense(activation='tanh',units=2,input_dim=1))
model1.add(Dense(1,activation='sigmoid'))
```

```python
model1.summary()
```

```python
model1.compile(loss='binary_crossentropy',optimizer='adadelta',metrics=['accuracy'])
```

```python
score = model1.evaluate(x5, x5, verbose=1)
print('test accuracy : ', score)
```

```python
hist=model1.fit(x5,x5,epochs=2000,verbose=1)
```

```python
from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot
 
SVG(model_to_dot(model1).create(prog='dot', format='svg'))
```

```python
classes = model1.predict(x5)
plt.plot(x5,classes,'.')
plt.show()
```

```python
import matplotlib.pyplot as plt
 
loss = hist.history['loss']
 
# lossのグラフ
plt.plot(range(len(loss)), loss, marker='.', label='loss')
plt.legend(loc='best', fontsize=10)
plt.grid()
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
 
acc = hist.history['acc']
val_acc = hist.history['val_acc']
 
# accuracyのグラフ
plt.plot(range(20), acc, marker='.', label='acc')
plt.plot(range(20), val_acc, marker='.', label='val_acc')
plt.legend(loc='best', fontsize=10)
plt.grid()
plt.xlabel('epoch')
plt.ylabel('acc')
plt.show()
```
